```python
lap_ant = session.laps.pick_drivers('ANT').pick_quicklaps().pick_fastest()
print(type(lap_ant))
print(lap_ant)
print(lap_ant['LapTime'])
print(type(lap_ant['LapTime']))
```

```python
telem_ant = lap_ant.get_telemetry()

print(telem_ant.columns.tolist())
print("ANT Shape:", telem_ant.shape)
print("ANT Track Distance:", telem_ant['Distance'].max())
print("ANT Max Speed:", telem_ant['Speed'].max())
print("ANT Min Speed:", telem_ant['Speed'].min())

# telem_ant.head()
# telem_ant.tail()
```

```python
lap_rus = session.laps.pick_drivers('RUS').pick_quicklaps().pick_fastest()

print("Antonelli's fastest lap:", lap_ant['LapTime'])
print("Russell's fastest lap:", lap_rus['LapTime'])

print(lap_ant['LapTime'].total_seconds() - lap_rus['LapTime'].total_seconds())

telem_rus = lap_rus.get_telemetry()

print(telem_rus.columns.tolist())
print("RUS Shape:", telem_rus.shape)
print("RUS Track Distance:", telem_rus['Distance'].max())
print("RUS Max Speed:", telem_rus['Speed'].max())
print("RUS Min Speed:", telem_rus['Speed'].min())

# telem_rus.head()
# telem_rus.tail()
```
